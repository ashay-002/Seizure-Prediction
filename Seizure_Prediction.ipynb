```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
eeg_data = x_resampled.values
labels = y_resampled.values
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(eeg_data)

scaled_data = scaler.transform(eeg_data)
```

```python
import numpy as np
import pandas as pd
import pywt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
def calculate_features(coeffs):
    features = []
    for i in range(len(coeffs)):
        energy = np.sum(coeffs[i] ** 2)
        entropy = -np.sum(np.abs(coeffs[i]) ** 2 * np.log(np.abs(coeffs[i]) ** 2))
        std_dev = np.std(coeffs[i])
        variance = np.var(coeffs[i])
        mean_abs = np.mean(np.abs(coeffs[i]))
        features.extend([energy, entropy, std_dev, variance, mean_abs])
    return features
```

```python
wavelet = 'db4'
level = 5

coefficients = []
for sample in scaled_data:
    coeffs = pywt.wavedec(sample, wavelet, level=level)
    features = calculate_features(coeffs)
    coefficients.append(features)


coefficients = np.array(coefficients)
```

```python
import pandas as pd


feature_names = ["energy_1", "entropy_1", "std_dev_1", "variance_1", "mean_abs_1",
                 "energy_2", "entropy_2", "std_dev_2", "variance_2", "mean_abs_2",
                 "energy_3", "entropy_3", "std_dev_3", "variance_3", "mean_abs_3",
                 "energy_4", "entropy_4", "std_dev_4", "variance_4", "mean_abs_4",
                 "energy_5", "entropy_5", "std_dev_5", "variance_5", "mean_abs_5",
                 "energy_6", "entropy_6", "std_dev_6", "variance_6", "mean_abs_6"]


coefficients_df = pd.DataFrame(coefficients, columns=feature_names)

print(coefficients_df)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(coefficients, labels, test_size=0.25, random_state=42)
```

Random Forest Classifier

```python
from sklearn.metrics import make_scorer, confusion_matrix, accuracy_score


def calculate_sensitivity(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sensitivity = tp / (tp + fn)
    return sensitivity

def calculate_specificity(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    specificity = tn / (tn + fp)
    return specificity


rf_classifier = RandomForestClassifier()


sensitivity_scorer = make_scorer(calculate_sensitivity)
specificity_scorer = make_scorer(calculate_specificity)
```

```python
from sklearn.model_selection import cross_validate

rf_classifier = RandomForestClassifier()

cv_results = cross_validate(rf_classifier, coefficients, labels, cv=10, scoring={'accuracy':'accuracy','sensitivity': sensitivity_scorer, 'specificity': specificity_scorer})

mean_accuracy = np.mean(cv_results['test_accuracy'])
mean_sensitivity = np.mean(cv_results['test_sensitivity'])
mean_specificity = np.mean(cv_results['test_specificity'])

print("Average Accuracy: {:.2f}".format(mean_accuracy))
print("Average Sensitivity: {:.2f}".format(mean_sensitivity))
print("Average Specificity: {:.2f}".format(mean_specificity))
```
